```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('Minitab.csv')
```

```python
data
```

```python
data=data.iloc[:,4:]
```

```python
data
```

```python
def augment_data(original_data, num_duplicates=10000, noise_factor=0.01):
    augmented_data = original_data.copy()
    for _ in range(num_duplicates):
        new_data_point = original_data.sample().values.flatten()
        noise = np.random.normal(0, noise_factor, size=new_data_point.shape)
        new_data_point += noise
        augmented_data = augmented_data.append(pd.Series(new_data_point, index=augmented_data.columns), ignore_index=True)
    return augmented_data
```

```python
augmented_data = augment_data(data, num_duplicates=10000)
```

```python
augmented_data = pd.concat([data, augmented_data], ignore_index=True)
```

```python
augmented_data
```

```python
X_augmented = augmented_data.drop(['Stress', 'Defomation'], axis=1)
y_augmented_stress = augmented_data['Stress']
y_augmented_deformation = augmented_data['Defomation']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_augmented, X_test_augmented, y_train_stress_augmented, y_test_stress_augmented, y_train_deformation_augmented, y_test_deformation_augmented = train_test_split(
    X_augmented, y_augmented_stress, y_augmented_deformation, test_size=0.2, random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled_augmented = scaler.fit_transform(X_train_augmented)
X_test_scaled_augmented = scaler.transform(X_test_augmented)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
model_stress_augmented = RandomForestRegressor()
model_stress_augmented.fit(X_train_scaled_augmented, y_train_stress_augmented)
```

```python
model_deformation_augmented = RandomForestRegressor()
model_deformation_augmented.fit(X_train_scaled_augmented, y_train_deformation_augmented)
```

```python
y_pred_stress_augmented = model_stress_augmented.predict(X_test_scaled_augmented)
y_pred_deformation_augmented = model_deformation_augmented.predict(X_test_scaled_augmented)
```

```python
mse_stress_augmented = mean_squared_error(y_test_stress_augmented, y_pred_stress_augmented)
r2_stress_augmented = r2_score(y_test_stress_augmented, y_pred_stress_augmented)
print(f'Stress Model (Augmented) - Mean Squared Error: {mse_stress_augmented}, R2 Score: {r2_stress_augmented}')
```

```python
mse_deformation_augmented = mean_squared_error(y_test_deformation_augmented, y_pred_deformation_augmented)
r2_deformation_augmented = r2_score(y_test_deformation_augmented, y_pred_deformation_augmented)
print(f'Deformation Model (Augmented) - Mean Squared Error: {mse_deformation_augmented}, R2 Score: {r2_deformation_augmented}')
```

```python
user_DOC = float(input("Enter DOC (mm): "))
user_displacement = float(input("Enter Displacement (mm): "))
user_rev = float(input("Enter Rev (rad/s): "))
```

```python
user_input = pd.DataFrame({
    'DOC (mm)': [user_DOC],
    'Displacement (mm)': [user_displacement],
    'Rev (rad/s)': [user_rev]
})
```

```python
user_input_scaled = scaler.transform(user_input)
```

```python
predicted_stress = model_stress_augmented.predict(user_input_scaled)[0]
predicted_deformation = model_deformation_augmented.predict(user_input_scaled)[0]
```

```python
print(f'Predicted Stress: {predicted_stress}')
print(f'Predicted Deformation: {predicted_deformation}')
```
