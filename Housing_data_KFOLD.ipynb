```python
import numpy as np
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data',header=None,delim_whitespace=True)
```

```python
df=df.values
```

```python
x=df[:,0:13]
y=df[:,13]
```

```python
def baseline():
  model = Sequential()
  model.add(Dense(13, input_shape=(13,), kernel_initializer='normal'))
  model.add(Dense(1, kernel_initializer='normal' ))
  model.compile(loss='mean_squared_error',optimizer='adam')
  return model
```

```python
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
estimator=KerasRegressor(baseline,epochs=50,batch_size=5)
kfold=KFold(n_splits=10)
results=cross_val_score(estimator,x,y,cv=kfold)
```

```python
print('Baseline: %2.2f (%2.2f) MSE' %(results.mean(),results.std()))
```
