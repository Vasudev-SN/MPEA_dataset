# Generate Duplicate Data using  'DATA AUGMENTATION'

```python
def augment_data(df, num_augmentations=5):
    augmented_data = []

    for _ in range(num_augmentations):
        # Randomly perturb the numerical values
        augmented_df = df.apply(lambda x: x + np.random.normal(scale=0.1) * x)

        # Append the augmented data to the list
        augmented_data.append(augmented_df)

    # Concatenate the original and augmented data
    augmented_df = pd.concat([df] + augmented_data, ignore_index=True)

    return augmented_df

# Specify the number of augmentations per data point
num_augmentations = 5
```

```python
import numpy as np
```

```python
augmented_df = augment_data(df, num_augmentations)
```

# Model Building using Sklearn

Random forests are versatile and can handle both linear and non-linear relationships in the data. They can capture complex patterns and interactions between features. Hence RandomRegressor was the chosen algorithm

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
features = augmented_df[['DOC (mm)', 'Displacement (mm)', 'Rev (rad/s)']]
target_stress = augmented_df['Stress']
target_deformation = augmented_df['Defomation']
```

```python
X_train, X_test, y_stress_train, y_stress_test, y_def_train, y_def_test = train_test_split(
    features, target_stress, target_deformation, test_size=0.2, random_state=42
)
```

```python
model_stress = RandomForestRegressor()
model_def = RandomForestRegressor()
```

```python
model_stress.fit(X_train, y_stress_train)
model_def.fit(X_train, y_def_train)
```

```python
X = augmented_df[['DOC (mm)', 'Displacement (mm)', 'Rev (rad/s)']]
y_stress = augmented_df['Stress']
y_deformation = augmented_df['Defomation']
```

```python
model_deformation = RandomForestRegressor(n_estimators=100, random_state=42)
model_deformation.fit(X, y_deformation)
```

```python
stress_predictions = model_stress.predict(X_test)
def_predictions = model_def.predict(X_test)
```

```python
print("Stress Model Metrics:")
print("MSE:", mean_squared_error(y_stress_test, stress_predictions))
print("MAE:", mean_absolute_error(y_stress_test, stress_predictions))
print("R-squared:", r2_score(y_stress_test, stress_predictions))

print("\nDeformation Model Metrics:")
print("MSE:", mean_squared_error(y_def_test, def_predictions))
print("MAE:", mean_absolute_error(y_def_test, def_predictions))
print("R-squared:", r2_score(y_def_test, def_predictions))
```
