```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X =iris.data
y = iris.target
```

```python
y =y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
y = OneHotEncoder(sparse=False).fit_transform(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
model1 = Sequential()
model1.add(Dense(150, input_dim = 4, activation='relu'))
model1.add(Dense(3, activation='relu'))
model1.add(Dense(3, activation='softmax'))
model1.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model1.summary()
```

```python
history1=model1.fit(x_train, y_train, validation_split=0.2, verbose =1, batch_size = 5, epochs = 5)
```

```python
model2 = Sequential()
model2.add(Dense(10, input_dim = 4, activation='relu'))
model2.add(Dense(10, activation='relu'))
model2.add(Dense(3, activation='softmax'))
model2.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model2.summary()
```

```python
history2=model2.fit(x_train, y_train, validation_split=0.2, verbose =1, batch_size = 5, epochs = 15)
```

```python
model3 = Sequential()
model3.add(Dense(10, input_dim = 4, activation='relu'))
model3.add(Dense(10, activation='relu'))
model3.add(Dense(10, activation='relu'))
model3.add(Dense(3, activation='softmax'))
model3.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history3=model3.fit(x_train, y_train, validation_split=0.2, verbose =1, batch_size = 5, epochs = 20)
```

```python
mse, mae = model3.evaluate(x_test, y_test)
print("MSE:", mse)
print("MAE:", mae)
```
